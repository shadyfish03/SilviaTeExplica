## **Función Query (paquete Pandas) Python**

---
Esta función nos permite filtras filas de dataframes basados en una expresión booleana (True, False) de manera más concisa. 



```python
data = pd.read_csv("https://raw.githubusercontent.com/shadyfish03/SilviaTeExplica/main/Salary_Data.csv")
```

```python
data.head()
```

```python
#1. Columna = Valor
data[data['Gender']=="Female"]
```

```python
data.query("Gender == 'Female'")
```

```python
#1.1 Columna tiene un espacio
data.query("`Job Title`=='HR Manager'")
```

```python
#1.2 Columna diferente a Valor
data.query("`Education Level`!='PhD'")
```

```python
#2. Columna = a varios valores
data.query("`Job Title` in ('Full Stack Engineer','Software Engineer')")
```

```python
#3. Columna >, < 
```

```python
data.query("150000 < Salary < 200000")
```

```python
#4. Condiciones en varias columnas
```

```python
data.query("Age < 40 and Salary > 200000")
```
